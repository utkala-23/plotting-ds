```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = sns.countplot(data=data, x='species')
ax.set_title("species count") 
plt.show()
```

```python
ax = sns.countplot(data=data,x='island')
ax.set_title("No. found per island")
plt.show()
```

Here we used the Countplot function of seaborn which basically counts the number of observations per category. The dataset used here is that of penguins so, the first plot shows the number of species of adele, chinstrap, gentoo respectively. Similarly, the second countplot shows the number of penguins found in each island.

```python
sns.barplot(x=data['species'].value_counts().index, y=data['species'].value_counts())
plt.show()
```

```python
def barplot(x, y, hue, data):
    fig, ax = plt.subplots(1, 1, figsize=(5, 3))
    sns.barplot(x, y, hue=hue, data=data)

for y in ['bill_length_mm','bill_depth_mm','flipper_length_mm']:
    barplot(x='species', y=y, hue='sex', data=data)
```

The above plotting was done using barplot. It divides the penguins into three categores of species: Adelie, Chinstrap species and Gentoo. It further divides each species type into sex of either male or female. It compares the bill length, bill depth, flipper length of penguins of each species with respect to their gender.

For example, we notice in the last graph that the flipper length of males in all three species is more than that of females of the species.

```python
def avg(x, y, hue, ax):
    sns.barplot(x, y, hue=hue, data=data, ax=ax)
    avg = data[y].sum().sum() / len(data)
    ax.axhline(avg, color='r')
    ax.set_title(f'The average {y} is {avg}')

fig, ax = plt.subplots(1, 3, figsize=(30, 10))
for idx, y in enumerate(['bill_length_mm','bill_depth_mm','flipper_length_mm']):
    avg('species', y, 'sex', ax[idx])
```

```python
def kdeplot(data, x, ax):
    sns.kdeplot(data[x], ax=ax)

fig, ax = plt.subplots(1, 2, figsize=(12, 7))
for idx, y in enumerate(['bill_length_mm','body_mass_g']):
    kdeplot(data, y, ax[idx])
```

Here, kdeplot is used to show the density of penguins in particular characteristics like ill length and body mass. We can see in the first plotting that the density of bill length is most between 45-50, this shows there are a large amount of penguins whose bill length lies in that range.

```python
#JOINTPLOT
sns.set({'figure.figsize': (15, 15)})
sns.jointplot('bill_length_mm', 'bill_depth_mm', data=data, kind='kde', space=1)
plt.show()
```
